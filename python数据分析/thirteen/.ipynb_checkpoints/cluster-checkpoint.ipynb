```python
data=pd.read_csv('data_bike.csv',encoding='gbk',index_col='日期与时间',parse_dates=True)
data
```

```python
pivoted=data.pivot_table('总数',index=data.index.date,columns=data.index.time)
pivoted
```

```python
total=pivoted.isnull().sum().sort_values(ascending=True)
missing_data=pd.DataFrame({'缺失值数量':total})
missing_data
```

```python
pivoted_fillnaed=pivoted.fillna(0)
```

```python
total2=pivoted_fillnaed.isnull().sum().sort_values(ascending=False)
missing_data2=pd.DataFrame({'缺失值数量':total2})
missing_data2
```
