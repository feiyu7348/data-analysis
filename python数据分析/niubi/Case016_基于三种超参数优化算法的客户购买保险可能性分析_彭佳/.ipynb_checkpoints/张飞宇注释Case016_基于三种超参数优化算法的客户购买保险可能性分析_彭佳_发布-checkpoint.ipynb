```python
train_features = np.array(train)
test_features = np.array(test)
train_features
```

```python
import lightgbm as lgb
from timeit import default_timer as timer 

model = lgb.LGBMClassifier()

start = timer()
model.fit(train_features, train_labels) 
train_time = timer() - start
```

```python
predictions=model.predict(test_features)
predictions
```

```python
predictions_1 = model.predict_proba(test_features)
predictions_1
```

```python
predictions_1[:,1]
```

```python
from sklearn.metrics import roc_auc_score 
auc = roc_auc_score(test_labels, predictions_1[:,1])


print('The baseline score on the test set is {:.4f}.'.format(auc))
print('The baseline training time is {:.4f} seconds'.format(train_time))
```

```python
import random  
from sklearn.model_selection import KFold  

MAX_EVALS = 80
N_FOLDS = 5 
```

```python
train_set = lgb.Dataset(train_features, train_labels
random_results = pd.DataFrame(columns = ['loss', 'params', 'evals', 'best_estimators', 'time'], 
                       index = list(range(MAX_EVALS)))
```

```python
type(train_set)
```

```python
def random_objective(params,evals, n_folds = N_FOLDS): 
  
    start = timer()
    
    cv_results = lgb.cv(params, train_set, num_boost_round = 500, nfold = n_folds, 
                        early_stopping_rounds = 70, metrics = 'auc', seed = 50)
    
    end = timer()
    best_score = np.max(cv_results['auc-mean'])
    loss = 1 - best_score
    
    best_estimators = int(np.argmax(cv_results['auc-mean']) + 1) 
    
    return [loss, params, evals, best_estimators, end - start]
```

```python
param_random = {
    'class_weight': [None, 'balanced'],
    'num_leaves': list(range(30, 150)),
    'learning_rate': list(np.logspace(np.log(0.005), np.log(0.2), base = np.exp(1), num = 1000)),
    'subsample_for_bin': list(range(20000, 300000, 20000)),
    'min_child_samples': list(range(20, 500, 5)),
    'reg_alpha': list(np.linspace(0, 1)),
    'reg_lambda': list(np.linspace(0, 1)),
    'colsample_bytree': list(np.linspace(0.6, 1, 10))
}

subsample_dist = list(np.linspace(0.5, 1, 100))
```

```python
params = {key: random.sample(value,2) for key, value in param_random.items()}
params
```

```python
%%capture 

random.seed(80)

for i in range(MAX_EVALS):
    
    params = {key: random.sample(value, 1)[0] for key, value in param_random.items()}
    print(params)
    
    if params['boosting_type'] == 'goss': 
        params['subsample'] = 1.0
    else:
        params['subsample'] = random.sample(subsample_dist, 1)[0]
        
        
    results_list = random_objective(params, i)
    random_results.loc[i, :] = results_list 
```

```python
random_results.sort_values('loss', ascending = True, inplace = True)
random_results.reset_index(inplace = True, drop = True)
random_results
```

```python
random_results.loc[0, 'params'] 
```

```python
best_random_params = random_results.loc[0, 'params'].copy()
best_random_estimators = int(random_results.loc[0, 'best_estimators'])
best_random_model = lgb.LGBMClassifier(best_estimators=best_random_estimators, n_jobs = -1, 
                                       objective = 'binary', **best_random_params, random_state = 50)
                                     
best_random_model.fit(train_features, train_labels)
```

```python
predictions_random = best_random_model.predict_proba(test_features)[:, 1]

print('The best model from random search scores {:.4f} on the test data.'.format(roc_auc_score(test_labels, predictions_random)))
```

```python
import csv      
from hyperopt import STATUS_OK 
from timeit import default_timer as timer

def bayes_objective(params, n_folds = N_FOLDS):

    global evals
    evals += 1  
    
    subsample = params['boosting_type'].get('subsample', 1.0)
    params['subsample'] = subsample
    params['boosting_type'] = params['boosting_type']['boosting_type']
     
    
   
    for parameter_name in ['num_leaves', 'subsample_for_bin', 'min_child_samples']:
        params[parameter_name] = int(params[parameter_name])
    
    start = timer()
    
    cv_results = lgb.cv(params, train_set, num_boost_round = 500, nfold = n_folds, 
                        early_stopping_rounds = 70, metrics = 'auc', seed = 50)
    
    run_time = timer() - start
    
   
    best_score = np.max(cv_results['auc-mean'])
    loss = 1 - best_score
    
    
    best_estimators = int(np.argmax(cv_results['auc-mean']) + 1)

   
    of_connection = open(out_file, 'a')
    writer = csv.writer(of_connection)
    writer.writerow([loss, params, evals, best_estimators, run_time]) 
    
    return {'loss': loss, 'params': params, 'evals': evals,
            'best_estimators': best_estimators, 
            'train_time': run_time, 'status': STATUS_OK}
```

```python
from hyperopt import hp 
```

```python
space = {
    'class_weight': hp.choice('class_weight', [None, 'balanced']),
    'boosting_type': hp.choice('boosting_type', [{'boosting_type': 'gbdt', 'subsample': hp.uniform('gdbt_subsample', 0.5, 1)}, 
                                                 {'boosting_type': 'dart', 'subsample': hp.uniform('dart_subsample', 0.5, 1)},
                                                 {'boosting_type': 'goss', 'subsample': 1.0}]),
    'num_leaves': hp.quniform('num_leaves', 30, 150, 1),
    'learning_rate': hp.loguniform('learning_rate', np.log(0.01), np.log(0.2)),
    'subsample_for_bin': hp.quniform('subsample_for_bin', 20000, 300000, 20000),
    'min_child_samples': hp.quniform('min_child_samples', 20, 500, 5),
    'reg_alpha': hp.uniform('reg_alpha', 0.0, 1.0),
    'reg_lambda': hp.uniform('reg_lambda', 0.0, 1.0),
    'colsample_bytree': hp.uniform('colsample_by_tree', 0.6, 1.0)
}
```

```python
import hyperopt.pyll.stochastic
from hyperopt.pyll.stochastic import sample 

print(hyperopt.pyll.stochastic.sample(space))
```

```python
from hyperopt import Trials
bayes_trials = Trials()
```

```python
out_file = "data/results_bayes_trials.csv"
of_connection = open(out_file, 'w')
writer = csv.writer(of_connection) 

writer.writerow(['loss', 'params', 'evals', 'best_estimators', 'train_time'])
of_connection.close()
```

```python
%%capture
from hyperopt import fmin
from hyperopt import tpe 
tpe_algorithm = tpe.suggest 

global  evals
evals = 0 


best = fmin(fn = bayes_objective, space = space, algo = tpe.suggest,
            max_evals = MAX_EVALS, trials = bayes_trials, rstate = np.random.RandomState(50)) 
```
